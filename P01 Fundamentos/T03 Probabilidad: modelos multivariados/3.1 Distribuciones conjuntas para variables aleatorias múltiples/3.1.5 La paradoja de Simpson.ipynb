# 3.1.5 La paradoja de Simpson

**Paradoja de Simpson:** $\quad$ se da cuando una tendencia que aparece en varios grupos de datos desaparece o cambia de signo al combinarlos

**Ejemplo:** $\quad$ en iris, $y$ decrece globalmente, pero crece intra-clase

```python
import numpy as np; import matplotlib.pyplot as plt
from sklearn.datasets import load_iris; from scipy.stats import linregress
X, y = load_iris(return_X_y=True)
fig, axs = plt.subplots(1, 2, figsize=(10,3), sharey=True); fig.tight_layout()
ax = axs[0]; ax.scatter(X[:,0], X[:,1]); I = np.array(ax.get_xlim())
s, i, R, _, _ = linregress(X[:, 0], X[:, 1])
ax.plot(I, s*I+i, label=f"R = {R:.2f}".format(R))
ax.legend(); ax = axs[1];
for c in (0, 1, 2):
    ax.scatter(X[np.where(y==c), 0], X[np.where(y==c), 1]); I = np.array(ax.get_xlim())
    s, i, R, _, _ = linregress(X[np.where(y==c), 0], X[np.where(y==c), 1])
    ax.plot(I, s*I+i, label=f"R = {R:.2f}".format(R))
ax.legend();
```

**Ejemplo:** $\quad$ la tasa de mortalidad de COVID-19 de Italia es menor que la de China en cada grupo de edad, pero mayor globalmente
