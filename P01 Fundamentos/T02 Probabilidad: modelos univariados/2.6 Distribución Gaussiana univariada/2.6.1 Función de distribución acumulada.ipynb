# 2.6.1 Función de distribución acumulada

**Función de distribución acumulada (cdf) Gaussiana:** $\displaystyle\quad\Phi(y;\mu,\sigma^2)=\int_{-\infty}^y \mathcal{N}(z\mid\mu,\sigma^2)\,dz\quad$ con $\;z=(y-\mu)/\sigma$

**Cdf usando la función error:** $\displaystyle\quad\Phi(y;\mu,\sigma^2)=\frac{1}{2}(1+\operatorname{erf}(z/\sqrt{2}))\quad$ donde $\quad\displaystyle\operatorname{erf}(u)=\frac{2}{\sqrt{\pi}}\int_0^ue^{-t^2}\,dt$

```python
import numpy as np; import matplotlib.pyplot as plt; from scipy.stats import norm
Y = norm(0, 1); y = np.linspace(-3, 3, 200); fig = plt.subplots(figsize=(5,3))
plt.grid(); plt.xlim(-3, 3); plt.ylim(0, 1); plt.plot(y, Y.cdf(y));
```

**Función probit:** $\;$ cdf inversa o función cuantil de la cdf Gaussiana, $\,\Phi^{-1}(q)=\{y_q:\Phi(y_q)=q\},\,0<q<1$

```python
import numpy as np; np.set_printoptions(precision=2); import matplotlib.pyplot as plt; from scipy.stats import norm
Y = norm(0, 1); y = np.linspace(0, 1, 200); fig = plt.subplots(figsize=(5,3))
plt.grid(); plt.xlim(0, 1); plt.ylim(-3, 3); plt.plot(y, Y.ppf(y)); print(Y.ppf([.025, .975]))
```
