# 2.7.2 Distribución de Cauchy


**Distribución de Cauchy o Lorentz:** $\;$ Student con $\,\nu=1$
$$\mathcal{C}(x\mid\mu,\gamma)=\frac{1}{\gamma\pi}\left[1+\left(\frac{x-\mu}{\gamma}\right)^2\right]^{-1}$$

**Colas muy pesadas:** $\;95\%\,$ de $\,\mathcal{C}(x\mid 0,1)\,$ en $\,[-12.7,12.7]$

```python
import numpy as np; import matplotlib.pyplot as plt; from scipy.stats import cauchy
x = np.linspace(-13, 13, 200); X = cauchy(loc=0, scale=1)
fig = plt.subplots(figsize=(5,3)); plt.plot(x, X.pdf(x));
```


**Half Cauchy:** $\;$ Cauchy (con $\mu=0$) plegada sobre $\,\mathbb{R}^+$
$$\mathcal{C}_+(x\mid\gamma)=\frac{2}{\gamma\pi}\left[1+\left(\frac{x}{\gamma}\right)^2\right]^{-1}$$

**Uso de half Cauchy:** $\;$ modelado Bayesiano como distribución en $\mathbb{R}^+$ con densidad finita en origen y cola pesada


```python
import numpy as np; import matplotlib.pyplot as plt; from scipy.stats import halfcauchy
x = np.linspace(-13, 13, 200); X = halfcauchy(scale=1)
fig = plt.subplots(figsize=(5,3)); plt.plot(x, X.pdf(x));
```
