# 2.7.3 Distribución de Laplace

**Distribución de Laplace:** $\;$ pdf de **parámetro de localización** $\,\mu\,$ y **parámetro de escala** $\,b>0$
$$\operatorname{Lap}(y\mid \mu,b)=\frac{1}{2b}\exp\left(-\frac{\lvert y-\mu\rvert}{b}\right)$$

```python
import numpy as np; import matplotlib.pyplot as plt; from scipy.stats import t, norm, laplace
Y = norm(0, 1); x = np.linspace(-3, 3, 200); T = t(df=1, loc=0, scale=1); L = laplace(loc=0, scale=1 / (2**0.5))
fig, axs = plt.subplots(1, 2, figsize=(12,3))
ax = axs[0]; ax.grid(); ax.set_xlim(-3, 3); ax.set_ylim(0, .7)
ax.plot(x, Y.pdf(x), "k:", label="Gaussian")
ax.plot(x, T.pdf(x), "b--", label="Student\n" + r"$(\nu=1)$")
ax.plot(x, L.pdf(x), "r-", label="Laplace"); ax.legend()
ax = axs[1]; ax.grid(); ax.set_xlim(-3, 3); ax.set_ylim(-5, 0)
ax.plot(x, np.log(Y.pdf(x)), "k:", label="log Gaussian")
ax.plot(x, np.log(T.pdf(x)), "b--", label="log Student\n" + r"$(\nu=1)$")
ax.plot(x, np.log(L.pdf(x)), "r-", label="log Laplace"); ax.legend();
```

**Propiedades:**
* Media: $\;\mu$
* Moda: $\;\mu$
* Varianza: $\;2b^2$

**Uso de Laplace:** $\;$ regresión lineal robusta y dispersa
