# 2.7.6 Distribución empírica

**Distribución empírica:** $\;$ de un conjunto de $N$ muestras $\,\mathcal{D}=\{x^{(1)},\dotsc,x^{(N)}\},\,$ derivado de una distribución $p(X)$, $X\in\mathbb{R}$

* **"Pdf":** $\;N$ funciones delta (picos) centradas en las muestras, $\qquad\hat{p}_N(x)=\frac{1}{N}\sum_{n=1}^N\delta_{x^{(i)}}(x)$

* **Cdf:** $\;$ función escalonada con saltos de altura $1/N$ en cada muestra, $\qquad\hat{P}_N(x)=\frac{1}{N}\sum_{n=1}^N\mathbb{I}(x^{(i)}\leq x)$

```python
import numpy as np; import matplotlib.pyplot as plt; from scipy.stats import ecdf
D = [6.23, 5.58, 7.06, 6.42, 5.20]  # one-mile run times (minutes)
_, ax = plt.subplots(1, 1, figsize=(6,3)); ecdf(D).cdf.plot(ax);
```
