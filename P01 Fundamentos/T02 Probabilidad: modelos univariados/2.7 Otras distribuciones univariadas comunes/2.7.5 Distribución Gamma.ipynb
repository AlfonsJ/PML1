# 2.7.5 Distribuci칩n Gamma y casos particulares

## Distribuci칩n Gamma

**Distribuci칩n Gamma:** $\;$ pdf en $\,\mathbb{R}^+\,$ gobernada por dos par치metros positivos, **forma** $\,a>0,\,$ y **ratio** $\,b>0\,$ o **escala** $\,s=1/b$
$$\operatorname*{Ga}(x\mid \text{shape}=a, \text{rate}=b)=\frac{b^a}{\Gamma(a)}x^{a-1}e^{-xb}\qquad
\operatorname*{Ga}(x\mid \text{shape}=a, \text{scale}=s)=\frac{1}{s^a\Gamma(a)}x^{a-1}e^{-x/s}$$
* $a:\;$ moda en $0$ si $a\leq 1$; si no, se aleja de $0$
* $b:\;$ al aumentar $b$, se reduce la escala horizontal


```python
import numpy as np; import matplotlib.pyplot as plt; from scipy.stats import gamma
x = np.linspace(0, 2, 100); fig, axs = plt.subplots(1, 2, figsize=(12,3))
ax = axs[0]; ax.grid(); ax.set_xlim(0, 2); ax.set_ylim(0, 4); ax.set_title(r'$a=1$')
for a, b in zip([1.0, 1.0, 1.0, 1.0], [0.5, 1.0, 2.0, 4.0]):
    ax.plot(x, gamma.pdf(x, a, scale=1/b, loc=0), label=f'$a={a:.1f},b={b:.1f}$'.format(a, b))
ax.legend();
ax = axs[1]; ax.grid(); ax.set_xlim(0, 2); ax.set_ylim(0, 4); ax.set_title(r'$a=2$')
for a, b in zip([2.0, 2.0, 2.0, 2.0], [0.5, 1.0, 2.0, 4.0]):
    ax.plot(x, gamma.pdf(x, a, scale=1/b, loc=0), label=f'$a={a:.1f},b={b:.1f}$'.format(a, b))
ax.legend();
```

**Propiedades:** $\quad$ media $\;\dfrac{a}{b}\quad$ moda $\;\dfrac{a-1}{b}\quad$ varianza $\;\dfrac{a}{b^2}$

## Distribuci칩n exponencial

**Distribuci칩n exponencial:** $\;$ tiempo entre eventos consecutivos en un proceso de Poisson de **ratio (intensidad)** $\,\lambda>0$
$$\operatorname*{Expon}(x\mid\lambda)=\operatorname*{Ga}(x\mid\text{shape}=1, \text{rate}=\lambda)$$

```python
import numpy as np; import matplotlib.pyplot as plt; from scipy.stats import expon
x = np.linspace(0, 2, 100); _, ax = plt.subplots(1, 1, figsize=(6,3))
ax.grid(); ax.set_xlim(0, 2); ax.set_ylim(0, 2);
[ ax.plot(x, expon.pdf(x, scale=1/b, loc=0), label=f'$\lambda={b:.2f}$'.format(b)) for b in [0.5, 1.0, 1.5] ]
ax.legend();
```

## Distribuci칩n Chi-cuadrado

**Distribuci칩n Chi-cuadrado:** $\;$ con $\nu$ **grados de libertad** es la distribuci칩n de la suma de cuadrados de $\nu$ normales 01, $\,S=\sum_{i=1}^{\nu}Z_i^2,\,$ con $\,Z_i\sim\mathcal{N}(0,1)$
$$\chi_{\nu}^2(x)=\operatorname*{Ga}\left(x\,\bigg\vert\,\text{shape}=\frac{\nu}{2}, \text{rate}=\frac{1}{2}\right)$$

```python
import numpy as np; import matplotlib.pyplot as plt; from scipy.stats import chi2
x = np.linspace(0, 10, 100); _, ax = plt.subplots(1, 1, figsize=(6,3))
ax.grid(); ax.set_xlim(0, 10); ax.set_ylim(0, 0.5);
[ ax.plot(x, chi2.pdf(x, df=k, loc=0), label=r'$\nu={:.2f}$'.format(k)) for k in [1, 2, 5, 10] ]
ax.legend();
```

## Distribuci칩n Gamma inversa

**Distribuci칩n Gamma inversa:** $\quad$ si $\,X\sim\operatorname{Ga}(\text{shape}=a,\text{rate}=b)$, $\quad 1/X\sim\operatorname{IG}(\text{shape}=a,\text{scale}=b)$
$$\operatorname{IG}(\text{shape}=a,\text{scale}=b)=\frac{b^a}{\Gamma(a)}x^{-(a+1)}e^{-b/x}$$

```python
import numpy as np; import matplotlib.pyplot as plt; from scipy.stats import invgamma
x = np.linspace(0, 2, 100); fig, axs = plt.subplots(1, 2, figsize=(12,3))
ax = axs[0]; ax.grid(); ax.set_xlim(0, 2); ax.set_ylim(0, 4); ax.set_title(r'$a=1$')
for a, b in zip([1.0, 1.0, 1.0, 1.0], [0.5, 1.0, 2.0, 4.0]):
    ax.plot(x, invgamma.pdf(x, a, scale=1/b, loc=0), label=f'$a={a:.1f},b={b:.1f}$'.format(a, b))
ax.legend();
ax = axs[1]; ax.grid(); ax.set_xlim(0, 2); ax.set_ylim(0, 4); ax.set_title(r'$a=2$')
for a, b in zip([2.0, 2.0, 2.0, 2.0], [0.5, 1.0, 2.0, 4.0]):
    ax.plot(x, invgamma.pdf(x, a, scale=1/b, loc=0), label=f'$a={a:.1f},b={b:.1f}$'.format(a, b))
ax.legend();
```

**Propiedades:** $\qquad$ media $\;\dfrac{b}{a-1}\,$ si $\,a>1\qquad$ moda $\;\dfrac{b}{a+1}\qquad$ varianza $\;\dfrac{b^2}{(a-1)^2(a-2)}\,$ si $\,a>2$
