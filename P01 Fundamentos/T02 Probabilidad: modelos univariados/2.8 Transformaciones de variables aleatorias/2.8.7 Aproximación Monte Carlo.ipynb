# 2.8.7 Aproximación Monte Carlo

**Aproximación Monte Carlo:** $\;$ muestrea $\,p(\boldsymbol{x})\,$ para aproximar $\,p(\boldsymbol{y}),\,\boldsymbol{y}=f(\boldsymbol{x}),\,$ cuando no es fácil derivar $\,p(\boldsymbol{y})\,$ analíticamente

**Ejemplo:** $\;$ distribución de $\,y=x^2\,$ con $\,p(x)\sim\operatorname{Unif}(-1,1)$
$$P(Y\leq y)=P(X^2\leq y)=P(-\sqrt{y}<X\leq\sqrt{y})=\sqrt{y}\quad\to\quad p(y)=\frac{1}{2\sqrt{y}}$$


```python
import numpy as np; import matplotlib.pyplot as plt; from scipy.stats import uniform
y = np.linspace(0.01, 1, 100); py =  1 / (2 * np.sqrt(y))
N = 1000; X = uniform(loc=-1, scale=2).rvs(N); f = lambda x: x*x; Y = f(X)
_, ax = plt.subplots(1, 1, figsize=(6, 3)); ax.hist(Y, bins=30, density=True, histtype='step'); ax.plot(y, py);
```
