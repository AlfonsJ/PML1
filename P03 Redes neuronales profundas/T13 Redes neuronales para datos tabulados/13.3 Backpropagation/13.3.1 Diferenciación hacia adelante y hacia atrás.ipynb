## El modelo y su Jacobiana

**Modelo:** $\;$ transformaci칩n $\;\boldsymbol{f}:\mathbb{R}^n\to\mathbb{R}^m\;$ tal que $\,\boldsymbol{o}=\boldsymbol{f}(\boldsymbol{x}),\,$ $\,\boldsymbol{x}\in\mathbb{R}^n\,$ y $\,\boldsymbol{o}\in\mathbb{R}^m,\;$ que compone $K$ funciones $\;\{\boldsymbol{f}_k\}$

**Modelo con $K=4$:** $\quad\boldsymbol{f}=\boldsymbol{f}_4\circ\boldsymbol{f}_3\circ\boldsymbol{f}_2\circ\boldsymbol{f}_1\quad$ tal que $\quad\boldsymbol{o}=\boldsymbol{f}(\boldsymbol{x})=\boldsymbol{f}_4(\boldsymbol{f}_3(\boldsymbol{f}_2(\boldsymbol{f}_1(\boldsymbol{x}))))$
$$\begin{align*}
\boldsymbol{f}_1&:\mathbb{R}^n\to\mathbb{R}^{m_1}&&&\boldsymbol{x}_2&=\boldsymbol{f}_1(\boldsymbol{x})\\%
\boldsymbol{f}_2&:\mathbb{R}^{m_1}\to\mathbb{R}^{m_3}&&&\boldsymbol{x}_3&=\boldsymbol{f}_2(\boldsymbol{x}_2)\\%
\boldsymbol{f}_3&:\mathbb{R}^{m_2}\to\mathbb{R}^{m_3}&&&\boldsymbol{x}_4&=\boldsymbol{f}_3(\boldsymbol{x}_3)\\%
\boldsymbol{f}_4&:\mathbb{R}^{m_3}\to\mathbb{R}^m&&&\boldsymbol{o}&=\boldsymbol{f}_4(\boldsymbol{x}_4)%
\end{align*}$$

**Jacobiana de $\boldsymbol{f}$:** $\;$ producto de las Jacobianas de $\boldsymbol{f}_1$, $\boldsymbol{f}_2$, $\boldsymbol{f}_3$ y $\boldsymbol{f}_4,\quad$ $\displaystyle\mathbf{J}_{\boldsymbol{f}}(\boldsymbol{x})=\frac{\partial\boldsymbol{o}}{\partial\boldsymbol{x}}\in\mathbf{R}^{m\times n}$
$$\frac{\partial\boldsymbol{o}}{\partial\boldsymbol{x}}%
=\frac{\partial\boldsymbol{o}}{\partial\boldsymbol{x}_4}%
\,\frac{\partial\boldsymbol{x}_4}{\partial\boldsymbol{x}_3}%
\,\frac{\partial\boldsymbol{x}_3}{\partial\boldsymbol{x}_2}%
\,\frac{\partial\boldsymbol{x}_2}{\partial\boldsymbol{x}}%
=\frac{\partial\boldsymbol{f}_4(\boldsymbol{x}_4)}{\partial\boldsymbol{x}_4}%
\,\frac{\partial\boldsymbol{f}_3(\boldsymbol{x}_3)}{\partial\boldsymbol{x}_3}%
\,\frac{\partial\boldsymbol{f}_2(\boldsymbol{x}_2)}{\partial\boldsymbol{x}_2}%
\,\frac{\partial\boldsymbol{f}_1(\boldsymbol{x})}{\partial\boldsymbol{x}}%
=\mathbf{J}_{\boldsymbol{f}_4}(\boldsymbol{x}_4)\,\mathbf{J}_{\boldsymbol{f}_3}(\boldsymbol{x}_3)\,\mathbf{J}_{\boldsymbol{f}_2}(\boldsymbol{x}_2)\,\mathbf{J}_{\boldsymbol{f}_1}(\boldsymbol{x})$$

**Recordatorio sobre la Jacobiana de $\boldsymbol{f}$:** $\quad\mathbf{J}_{\boldsymbol{f}}(\boldsymbol{x})\in\mathbf{R}^{m\times n}$
$$\mathbf{J}_{\boldsymbol{f}}(\boldsymbol{x})%
=\frac{\partial\boldsymbol{f}(\boldsymbol{x})}{\partial\boldsymbol{x}}%
=\begin{bmatrix}%
\frac{\partial f_1}{\partial x_1}&\cdots&\frac{\partial f_1}{\partial x_n}\\%
\vdots&\ddots&\vdots\\%
\frac{\partial f_m}{\partial x_1}&\cdots&\frac{\partial f_m}{\partial x_n}%
\end{bmatrix}%
=\begin{bmatrix}
\mathbf{\nabla} f_1(\boldsymbol{x})^t\\%
\vdots\\%
\mathbf{\nabla} f_m(\boldsymbol{x})^t%
\end{bmatrix}%
=\begin{bmatrix}
\frac{\partial\boldsymbol{f}}{\partial x_1},\ldots,\frac{\partial\boldsymbol{f}}{\partial x_n}%
\end{bmatrix}$$
* **La Jacobiana de un campo escalar ($m=1$) es su gradiente transpuesto:** $\;\mathbf{J}_f(\boldsymbol{x})=\mathbf{\nabla} f(\boldsymbol{x})^t$
