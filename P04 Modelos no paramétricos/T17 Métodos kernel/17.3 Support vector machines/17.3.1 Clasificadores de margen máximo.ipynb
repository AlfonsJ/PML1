# 17.3.1 Clasificadores de margen máximo

**Clasificador de referencia:** $\;$ $\;$ binario, para etiquetas $\tilde{y}\in\{-1,+1\}$
$$h(\boldsymbol{x})=\operatorname{sgn}(f(\boldsymbol{x}))%
\quad\text{con}\quad%
f(\boldsymbol{x})=\boldsymbol{w}^t\boldsymbol{x}+w_0$$

**Clasificador margen máximo:** $\;$ de máxima distancia entre la frontera y su muestra más cercana (y que clasifique bien)
<div align="center"><img src="Figure_17.12.png" width="500"/></div>

**Distancia $r$ de un punto $\boldsymbol{x}$ a la frontera:**
<div><table border-collapse: collapse><tr>
<td style="border: none; text-align:left; vertical-align:top; padding:0; margin:0;" width=350>

$$\boldsymbol{x}=\boldsymbol{x}_{\bot}+r\frac{\boldsymbol{w}}{\lVert{\boldsymbol{w}}\rVert}$$

$$\begin{align*}
f(\boldsymbol{x})&=\boldsymbol{w}^t\boldsymbol{x}+w_0\\%
&=(\boldsymbol{w}^t\boldsymbol{x}_{\bot}+w_0)+r\frac{\boldsymbol{w}^t\boldsymbol{w}}{\lVert{\boldsymbol{w}}\rVert}\\%
&=r\,\lVert{\boldsymbol{w}}\rVert\quad\to\quad r=\frac{f(\boldsymbol{x})}{\lVert{\boldsymbol{w}}\rVert}%
\end{align*}$$

Invarianza a escalado de parámetros: $\;k>0$
$$\frac{(k\boldsymbol{w})^t\boldsymbol{x}+kw_0}{\lVert{k\boldsymbol{w}}\rVert}
=\frac{k(\boldsymbol{w}^t\boldsymbol{x}+w_0)}{k\lVert{\boldsymbol{w}}\rVert}=r$$

**Problema primal:** $\;$ hallar un clasificador de margen máximo y sin error para $N$ datos, $\{(\boldsymbol{x}_n,\tilde{y}_n)\}$
$$\begin{align*}
\boldsymbol{w}^*,w_0^*%
&=\operatorname*{argmax}_{\boldsymbol{w},w_0}\;%
\frac{1}{\lVert{\boldsymbol{w}}\rVert}\min_n\;\tilde{y}_nf_n%
&&\text{sujeto a}\;\{\tilde{y}_nf_n>0\}\\
&=\operatorname*{argmax}_{\boldsymbol{w},w_0}\;%
\frac{1}{\lVert{\boldsymbol{w}}\rVert}%
&&\text{sujeto a}\;\{\tilde{y}_nf_n\geq 1\}%
\;\;(\text{escalando tal que}\;\min_n\;\tilde{y}_nf_n=1)\\%
&=\operatorname*{argmax}_{\boldsymbol{w},w_0}\;%
\frac{2}{\lVert{\boldsymbol{w}}\rVert}%
&&\text{sujeto a}\;\{\tilde{y}_nf_n\geq 1\}\\%
&=\operatorname*{argmin}_{\boldsymbol{w},w_0}\;%
\frac{1}{2}\lVert{\boldsymbol{w}}\rVert^2%
&&\text{sujeto a}\;\{\tilde{y}_nf_n\geq 1\}
\end{align*}$$

**Escalado de datos:** $\;$ conviene escalar los datos antes de hallar $\boldsymbol{w}^*,w_0^*$
